# Life Insurance

Life insurance is a financial guarantee provided to beneficiaries, who are usually family members, in the event of serious illness, disability, or the death of the insured, also known as the policyholder.

This type of insurance, like others, is established through a contract with an insurance company. Being a contract, there are different types of life insurance, with varying periods of validity, the number of policyholders, and other characteristics that, depending on the insurance company, can be customized.

The operation of this insurance is based on payments, usually made monthly (but they can take other forms such as semi-annual or even a lump-sum payment), by the policyholder. In the event of a claim, i.e., any of the situations covered by the insurance, the insurance company is contacted and pays compensation to the beneficiaries specified in the contract.

For this reason, life insurance is often used as a financial planning instrument, providing security to the family of the policyholder.

The amount to be paid for this type of contract is variable and should take into account factors such as: the coverage period of the insurance, the number of people who will receive the benefit, types of coverage (serious illness, temporary disability, disability, death), family history of illnesses, the profession of the policyholder (as there are professions considered riskier, such as police officers and firefighters, for example), among other factors.

To address this issue for insurance companies, it is possible to create a machine-learning model that uses a regression algorithm to predict the amount that should be charged to the policyholder.

## 3. Model Adjustment

The model will be optimized using the R2 metric, and as it is already the default for the `tune_model` function, there is no need to specify it.

Just as a reminder, R2 or R², the Coefficient of Determination, is a way to measure the accuracy of the model. However, it does not deal with errors themselves but with variances concerning the mean and the line. In this way, it measures how well the model fits the data.

Its results range from 0 to 1.0, with the closer it is to 1.0, the better the result.

Its formula is given by:

$$
R² (y, \widehat{y})=1-\frac{\sum_{i=1}^{N}(y_i - \widehat{y}_i)²}{\sum_{i=1}^{N}(y_i - \bar{y}_i)²}
$$


```python
# parameter adjustment
tuned_gbr = tune_model(gbr)
```

With the adjustment of parameters, we had a small improvement in the R2 result: we went from 0.8503 to 0.8514.
